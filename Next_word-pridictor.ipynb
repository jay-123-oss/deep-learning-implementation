```python
import re

def clean_text(text):
    text = text.lower()

    # unicode weird chars remove (like � etc.)
    text = text.encode('ascii', 'ignore').decode()

    # keep only letters, numbers, space, dot
    text = re.sub(r'[^a-zA-Z0-9\s\.]', '', text)

    # extra spaces fix
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
text = clean_text(text)
print(text)
```

```python
# !pip install tensorflow
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
# tokenizer.index_word
```

```python
len(tokenizer.index_word)
```

```python
input_sequences = []
# Split the text into sentences using the dot as a delimiter.
# Filter out any empty strings that might result from the split.
sentences = [s.strip() for s in text.split('.') if s.strip()]

for sentence in sentences:
  # Tokenize each sentence into a sequence of numbers
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  # Generate input sequences for training. Each sequence will be a prefix
  # of the tokenized sentence, with the last word being the target.
  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
input_sequences
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences.shape
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
X
```

```python
X.shape
```

```python
y = padded_input_sequences[:,-1]
```

```python
y.shape
```

```python
import numpy as np
from sklearn.utils import shuffle
from tensorflow.keras.utils import to_categorical

x_small = np.array(X[:1000])
# Create y_small from the original integer labels (padded_input_sequences[:,-1])
y_small_labels = np.array(padded_input_sequences[:1000, -1])

# Convert y_small_labels to one-hot encoding
vocab_size = len(tokenizer.index_word) + 1
y_small = to_categorical(y_small_labels, num_classes=vocab_size)

# Shuffle x_small and the one-hot encoded y_small
x_small, y_small = shuffle(x_small, y_small, random_state=42)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=len(tokenizer.index_word) + 1)
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, BatchNormalization

vocab_size = len(tokenizer.index_word) + 1
num_classes = vocab_size

model = Sequential()

# Embedding (very important for text)
model.add(Embedding(input_dim=vocab_size, output_dim=128))

# Hidden Layer 1 (strong but controlled)
model.add(LSTM(128, return_sequences=True))
model.add(Dropout(0.3))
model.add(BatchNormalization())

# Hidden Layer 2
model.add(LSTM(64))
model.add(Dropout(0.3))
model.add(BatchNormalization())

# Dense layer
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))

# Output layer
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-5
)
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Compile the model with an optimizer, loss function, and metrics
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    x_small, y_small,
    epochs=10,
    batch_size=256,
    callbacks=[lr_scheduler, early_stop]
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['loss'], label='Loss')

plt.title('Accuracy & Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Value')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Evaluate the model on the training data
loss, accuracy = model.evaluate(X, y, verbose=0)
print(f'Training Loss: {loss:.4f}')
print(f'Training Accuracy: {accuracy:.4f}')
```
